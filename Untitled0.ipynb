```python
import face_recognition
import os

known_face_encodings = []
known_face_names = []

KNOWN_FACES_DIR = "known_faces"

for name in os.listdir(KNOWN_FACES_DIR):
    person_folder = os.path.join(KNOWN_FACES_DIR, name)
    if not os.path.isdir(person_folder):
        continue

    for filename in os.listdir(person_folder):
        image_path = os.path.join(person_folder, filename)
        image = face_recognition.load_image_file(image_path)

        encodings = face_recognition.face_encodings(image)
        if encodings:
            known_face_encodings.append(encodings[0])
            known_face_names.append(name)

print("Encoding completed ✅")
print("Total faces:", len(known_face_names))
```

```python
from google.colab import files
import face_recognition
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Upload test image
uploaded = files.upload()
test_image_path = list(uploaded.keys())[0]

# Load image
test_image = face_recognition.load_image_file(test_image_path)
face_locations = face_recognition.face_locations(test_image)
face_encodings = face_recognition.face_encodings(test_image, face_locations)

image_bgr = cv2.cvtColor(test_image, cv2.COLOR_RGB2BGR)

results = []  # to store final output text

for idx, ((top, right, bottom, left), face_encoding) in enumerate(
    zip(face_locations, face_encodings), start=1):

    distances = face_recognition.face_distance(known_face_encodings, face_encoding)
    best_match_index = np.argmin(distances)
    best_distance = distances[best_match_index]

    confidence = max(0, (1 - best_distance)) * 100

    if best_distance < 0.6:
        name = known_face_names[best_match_index]
    else:
        name = "Unknown"

    # Draw ONLY green box (no text)
    cv2.rectangle(
        image_bgr,
        (left, top),
        (right, bottom),
        (0, 255, 0),
        thickness=6
    )

    results.append(f"Face {idx}: {name} — Confidence: {confidence:.2f}%")

# Show image
plt.figure(figsize=(6,10))
plt.imshow(cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()

# 🟢 PRINT RESULTS BELOW IMAGE
print("🔍 Face Recognition Results:")
for r in results:
    print(r)
```
